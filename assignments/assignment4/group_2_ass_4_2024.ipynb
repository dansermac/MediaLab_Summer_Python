### 5. Generate a function named as calling_data that take as input the `name of the firm  or list of names (firm_name)` and `(start_year, end_year)`. The function should return a tuple with a dataframe with the same format as the one resulting from task 1-4, and your dictionary of columns. The function should work without `end_year`. It means it only takes `start_year`.

```python
# Reshape database to long format, using data frame with short names.
long_bogota_database = bogota_database.melt(id_vars=['Company Name'], 
                           var_name='Year', 
                           value_name='LCU')
```

```python
# Change data type in "Year" from str to int.
long_bogota_database['Year'] = long_bogota_database['Year'].astype(int)
```

```python
#sort values for better visualization of the data.
new_data = long_bogota_database.sort_values (by = ['Company Name', 'Year'])
new_data
```

```python
#Creating the required function.
def calling_data (firm_name, start_year, end_year = None):
     
    
    filtered_new_data = new_data[(new_data['Company Name'].isin (firm_name)) & (new_data['Year'] >= start_year)]
     
        
    return filtered_new_data
```

```python
# Using the "calling_data" function with just a company name.
calling_data (["DINASTIA INVERSIONES S A"], 2011)
```
