## Exercise 2

#### Delete columns in Dollars

```python
dollar_columns = [col for col in data.columns if 'USD' in col]
data_without_dollars = data.drop(columns=dollar_columns)
```

#### Maintain columns beginning with LCU 2023 or Company name Latin alphabet

```python
columns_to_keep = [col for col in data_without_dollars.columns if col == 'Operating revenue (Turnover)\nth LCU 2023' or col == 'Company name Latin alphabet']
data_selected_columns = data_without_dollars[columns_to_keep]
```

#### Changing the format to long

```python
data_long_format = data_selected_columns.melt(id_vars=['Company name Latin alphabet'], var_name='Year', value_name='LCU')
```

#### Extracting only the year in the column 'Year'.

```python
data_long_format['Year'] = data_long_format['Year'].str.extract('(\d{4})') 
```
