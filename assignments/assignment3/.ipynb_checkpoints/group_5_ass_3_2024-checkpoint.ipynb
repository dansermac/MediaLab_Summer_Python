## Exercise 2

#### Delete columns in Dollars

```python
dollar_columns = [col for col in data.columns if 'USD' in col]
data_without_dollars = data.drop(columns=dollar_columns)
```

#### Maintain columns beginning with LCU 2023 or Company name Latin alphabet

```python
columns_to_keep = [col for col in data_without_dollars.columns if col == 'Operating revenue (Turnover)\nth LCU 2023' or col == 'Company name Latin alphabet']
data_selected_columns = data_without_dollars[columns_to_keep]
```

#### Changing the format to long

```python
data_long_format = data_selected_columns.melt(id_vars=['Company name Latin alphabet'], var_name='Year', value_name='LCU')
```

#### Extracting only the year in the column 'Year'.

```python
data_long_format['Year'] = data_long_format['Year'].str.extract('(\d{4})') 
```

## Exercise 5

```python
# Replication of the long data using for loop

replicated_data = []

for index, row in data_long_format.iterrows():

    for year_col in data_long_format.columns:

        replicated_data.append({
            'Company': row['Company name Latin alphabet'],
            'Year'   : row['Year'],
            'LCU'    : row['LCU']
        })


# Transform the list in a DataFrame
replicated_long_df = pd.DataFrame(replicated_data)
replicated_long_df.head()
```

## Exercise 6

```python
# Replication of the long data, but instead of using methods of pandas, using while loop.
```

```python
replicated_data = []
row_index = 0

while row_index < len(data_long_format):

    row = data_long_format.iloc[row_index]

    for year_col in data_long_format.columns:

        replicated_data.append({
            'Company': row['Company name Latin alphabet'],
            'Year'   : row['Year'],
            'LCU'    : row['LCU']
        })

    row_index += 1

# Transform the list in a DataFrame
replicated_long_df_while = pd.DataFrame(replicated_data)
replicated_long_df_while.head()
```
